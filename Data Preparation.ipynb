## Resume Data

```python
import pickle
```

```python
resume1 = pickle.load(open('data/train_data.pkl','rb'))
```

```python
len(resume1)
```

```python
resume1[0][0]
```

```python
resume_data = pd.Series(resume1[0][0], name='Resume')
for i in range(1, len(resume1)):
    res = pd.Series(resume1[i][0], name='Resume')
    resume_data = pd.concat([resume_data, res], ignore_index=True)
resume_data
```

```python
resume_data = resume_data.to_frame()
resume_data
```

```python
# check how my second resume data looks like
open('data/CVs/cv1', "rb").read().decode('utf-8').replace("\n", "")
```

```python
for i in range(1,251):
    if os.path.exists('data/CVs/cv{}'.format(i))==True:
        resume_data.loc[len(resume_data.index)] = open('data/CVs/cv{}'.format(i), "rb").read().decode('utf-8').replace("\n", "")
    else:
        continue
```
