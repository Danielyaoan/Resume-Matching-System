### Randomly select 2000 jobs for prototype demonstration.

```python
job_match = []
randomlist = random.sample(range(285429), 2000)
randomlist
```

```python
openai.api_key = "Your-openai-api-key"
def Job_summary(randomlist, job_data):
    prompt = "Extract education and key requirements from this job description, ignore salary range and limit the number of words to 150: "
    for i in range(len(randomlist)):
        rnd_index = randomlist[i]
        # avoid over max tokens
        while len(str(job_data.iloc[[rnd_index]]['FullDescription']))>=3500:
            rnd_index+=1
        job_text = str(job_data.iloc[[rnd_index]]['FullDescription'])

        response = openai.Completion.create(
            engine="text-davinci-003",
            prompt=prompt+job_text,
            temperature=0.0,
            max_tokens=400,
            top_p=1,
            frequency_penalty=1.0,
            presence_penalty=1.2,
            stop=None,
            n=1,
        )
        section = response.choices[0].text.strip()
        # extract education and requirements parts
        education = ""
        key_requirements = ""
        education_index = section.index('Education')
        key_requirements_index = section.index('Key Requirements')

        education += section[education_index+10:key_requirements_index-1].strip()
        key_requirements += section[key_requirements_index+17:len(section)].strip()
        # prepare for data frame
        temporary = list((job_data.iloc[[rnd_index]]['Title'], job_data.iloc[[rnd_index]]['FullDescription'], education, key_requirements))
        job_match.append(temporary)
    return job_match
```

```python
# summarize separately to check if the outcome are acceptable and avoid server crash
#job_match = Job_summary(randomlist[:500], job_data)
#job_match = Job_summary(randomlist[500:1000], job_data)
#job_match = Job_summary(randomlist[1000:1500], job_data)
job_match = Job_summary(randomlist[1500:2000], job_data)
```
