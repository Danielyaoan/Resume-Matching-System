## An alternative way to compute similarity

```python
openai.api_key = "Your-openai-api-key"

prompt = "Based on the resume and job description below, generate similar words for skills, experience and requirements, compare those words strictly and give me the similarity, if resume similar words of skills and experience satisfy all requirements in job description, similarity is 100; if resume satisfy no requirements in job description, similarity is 0: "
job_text = str(df_match_job.iloc[[0]]['Summarized'])
resume_text = str(df_match_resume.iloc[[9]]['Extracted'])

response = openai.Completion.create(
    engine="text-davinci-003",
    prompt=prompt+'resume: '+resume_text+'job description: '+job_text,
    temperature=0.0,
    max_tokens=400,
    top_p=1,
    frequency_penalty=1.0,
    presence_penalty=1.2,
    stop=None,
    n=1,
)
```

```python
section = response.choices[0].text.strip()
print(section)
```
